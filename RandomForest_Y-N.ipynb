```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

#load train and test data
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('validation.csv')

#check class distribution of answers in train and test data
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.countplot(x='answer', data=train_data, ax=ax[0])
sns.countplot(x='answer', data=test_data, ax=ax[1])
ax[0].set_title('Train Data Class Distribution')
ax[1].set_title('Test Data Class Distribution')
plt.show()

#preprocess train data
vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(train_data['question'] + ' ' + train_data['passage'])
y_train = train_data['answer']

#train logistic regression model
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train_count, y_train)

#preprocess test data
X_test_count = vectorizer.transform(test_data['question'] + ' ' + test_data['passage'])
y_test = test_data['answer']

#evaluate the model on test data
accuracy = clf.score(X_test_count, y_test)
print(f'Test accuracy: {accuracy}')

#visualize the feature importance
#visualize the feature importance
feature_names = vectorizer.get_feature_names()
coefs = pd.DataFrame({'feature_name': feature_names, 'coef': clf.coef_.flatten()})
top_features = coefs.sort_values('coef', ascending=False).head(20)
bottom_features = coefs.sort_values('coef', ascending=True).head(20)
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(x='coef', y='feature_name', data=top_features, ax=ax[0])
sns.barplot(x='coef', y='feature_name', data=bottom_features, ax=ax[1])
ax[0].set_title('Top 20 Features')
ax[1].set_title('Bottom 20 Features')
plt.show()
```

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

accuracies = []
runtimes = []

train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('validation.csv')

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.countplot(x='answer', data=train_data, ax=ax[0])
sns.countplot(x='answer', data=test_data, ax=ax[1])
ax[0].set_title('Train Data Class Distribution')
ax[1].set_title('Test Data Class Distribution')
plt.show()

vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(train_data['question'] + ' ' + train_data['passage'])
y_train = train_data['answer']

for n in [1, 2, 3, 4, 5, 10, 40]:
#start time
    start_time = time.time()

# train random forest classifier
    clf = RandomForestClassifier(n_estimators=n, random_state=42)
    clf.fit(X_train_tfidf, y_train)

# preprocess test data
    X_test_tfidf = vectorizer.transform(test_data['question'] + ' ' + test_data['passage'])
    y_test = test_data['answer']

# evaluate the model on test data
    y_pred = clf.predict(X_test_tfidf)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

#end time
    end_time =time.time()
    runtime = end_time - start_time
    runtimes.append(runtime)
    
    print("n_estimators : "+ str(n) +" __ run time : "+str(runtime)+" __ accuracy : "+ str(accuracy))

fig, ax = plt.subplots(2, 1, figsize=(10, 8))
ax[0].plot([1, 2, 3, 4, 5, 10, 40], accuracies, marker='o')
ax[0].set_xlabel('n_estimators')
ax[0].set_ylabel('Accuracy')
ax[1].plot([1, 2, 3, 4, 5, 10, 40], runtimes, marker='o')
ax[1].set_xlabel('n_estimators')
ax[1].set_ylabel('Runtime (seconds)')
plt.show()
```
